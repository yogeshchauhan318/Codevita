**Missing data**

```python
df["designation"]=df["designation"].isna().dropna()
df["taster_name"]=df["taster_name"].isna().dropna()
df["taster_twitter_handle"]=df["taster_twitter_handle"].isna().dropna()
```

**Visualization**

```python
plt.figure(figsize=(25,10))
sns.countplot(df.country,order = df.country.value_counts().index[:10])
plt.title("Counts of the Evaluation of Countries")
plt.xlabel("Countries")
plt.ylabel("Counts")
plt.show()
```

```python
df = df.drop(["Unnamed: 0", "designation", "region_1", "region_2", "taster_name", "taster_twitter_handle"], axis=1)
df.head()
```

```python
df = df.dropna(subset=["country","price","province","variety"])
```

```python
plt.hist(data=df, x='points', bins=np.arange (80, df['points'].max()+1, 1))
plt.xlim(80,100)
plt.title('Dristribution of points')
plt.xlabel('Points')
plt.ylabel('Frequency');
```

```python
top10=df["country"].value_counts()[:10]
top10.plot(kind="bar")
plt.title("No. of wine ratings by countries")
plt.xlabel("Country")
plt.ylabel("No. of ratings")
```

```python
plt.figure(figsize=(20,20))
sns.boxplot(data=df,x="points",y="country")
plt.title("Points by country")
plt.xlabel("Points")
plt.ylabel("Country")
```

```python
average_price = df.groupby("country").agg("mean")
average_price = average_price.sort_values(by="price",ascending=False)
y = average_price.index.to_series()
x = average_price["price"]

plt.figure(figsize=(20,20))
sns.barplot(x=x,y=y)
plt.title("Average price by country")
plt.xlabel("Average price")
plt.ylabel("country")
```

```python
bins = np.arange(0,df["price"].max()+5,5)
plt.hist(data=df,x="price",bins=bins)
plt.xlim(0,150)
plt.title("Wine price")
plt.xlabel("Price")
plt.ylabel("Frequency")
```

```python
x=[1,3,10,30,100,300,1000,3000,10000]

plt.figure(figsize=(15,15))
sns.regplot(x="price",y="points",data=df,logx=True)
plt.title("Points vs Price",fontsize=15)
plt.xscale("log")
plt.xticks(x,x)
plt.xlabel("Price")
plt.ylabel("Points")
```
